```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#eligo columnas de sexocada base ---
def pick(df, candidates):
    for c in candidates:
        if c in df.columns: return c
    return None

sex05 = pick(df_T105_NOA, ["sexo","CH04","SEXO","ch04"])
sex25 = pick(df_T125_NOA, ["sexo","CH04","SEXO","ch04"])
w05   = pick(df_T105_NOA, ["pondera","PONDERA","peso"])
w25   = pick(df_T125_NOA, ["pondera","PONDERA","peso"])

#junto los dataset (sexo + año )
d05 = df_T105_NOA[[sex05] + ([w05] if w05 else [])].rename(columns={sex05:"sexo"})
d05["ano4"] = 2005
if w05: d05 = d05.rename(columns={w05:"w"})

d25 = df_T125_NOA[[sex25] + ([w25] if w25 else [])].rename(columns={sex25:"sexo"})
d25["ano4"] = 2025
if w25: d25 = d25.rename(columns={w25:"w"})

df = pd.concat([d05, d25], ignore_index=True)
if "w" not in df.columns: df["w"] = 1

# Normalizo sexo (formas en las que lo ponen)
map_sexo = {
    "1":"Varón","2":"Mujer",
    "m":"Varón","hombre":"Varón","varón":"Varón","varon":"Varón","masculino":"Varón",
    "f":"Mujer","mujer":"Mujer","femenino":"Mujer"
}
df["sexo_lbl"] = df["sexo"].astype(str).str.strip().str.lower().map(map_sexo)

#porcentaje por año
comp = (df.dropna(subset=["sexo_lbl"])
          .groupby(["ano4","sexo_lbl"], as_index=False)["w"].sum()
          .rename(columns={"w":"n"}))
comp["porcentaje"] = comp.groupby("ano4")["n"].transform(lambda s: 100*s/s.sum())

#gráfico 
plt.figure(figsize=(7,4))
sns.barplot(data=comp, x="ano4", y="porcentaje", hue="sexo_lbl", hue_order=["Mujer","Varón"])
plt.ylim(0,100)
plt.title("Composición por sexo — 2005 vs 2025")
plt.xlabel("Año"); plt.ylabel("Porcentaje")
plt.legend(title="Sexo")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# ================================
# 1) Partimos de tu base
# ================================
columnas_nuevas = [
    "codusu", "nro_hogar", "ano4",
    "sexo", "años_cumplidos", "estado_civil", "cobertura_medica",
    "nivel_ed", "estado", "inactividad", "ipcf"
]
df_hogar = df_interes[columnas_nuevas].copy()

# ================================
# 2) Tabla de adulto equivalente
# ================================
# El archivo trae encabezado sucio y una fila 'Edad' que hay que sacar
tabla = (pd.read_excel("tabla_adulto_equiv.xlsx", skiprows=2)
           .rename(columns={"Unnamed: 0":"edad_label","Unnamed: 1":"Mujer","Unnamed: 2":"Varón"}))

# eliminar filas sin etiqueta y la fila de encabezado 'Edad'
tabla = tabla.dropna(subset=["edad_label"]).copy()
tabla["edad_label"] = tabla["edad_label"].astype(str).str.strip()
tabla = tabla[tabla["edad_label"] != "Edad"]

# formato largo
tabla_long = tabla.melt(
    id_vars="edad_label",
    value_vars=["Mujer", "Varón"],
    var_name="sexo_tabla",
    value_name="adulto_equiv"
)
tabla_long["adulto_equiv"] = pd.to_numeric(tabla_long["adulto_equiv"], errors="coerce")

# ================================
# 3) Normalizadores
# ================================
def normalizar_sexo_valor(x):
    # acepta 1/2, 1.0/2.0, varon/varón/hombre/m, mujer/f, etc.
    try:
        xn = float(x)
        if int(xn) == 1: return "Varón"
        if int(xn) == 2: return "Mujer"
    except Exception:
        pass
    s = str(x).strip().lower()
    if s in {"1","1.0","varon","varón","hombre","masculino","m","v"}: return "Varón"
    if s in {"2","2.0","mujer","femenino","f"}:                       return "Mujer"
    return pd.NA

def edad_a_tramo(e):
    # OJO: las etiquetas deben coincidir EXACTO con la tabla (p.ej., "1año" SIN espacio)
    if pd.isna(e): return pd.NA
    try:
        e = int(np.floor(float(e)))
    except Exception:
        return pd.NA
    if e < 1:            return "Menor de 1 año"
    if e == 1:           return "1año"           # <-- cambio clave: SIN espacio
    if 2  <= e <= 14:    return f"{e} años"
    if e == 15:          return "15 años"
    if e == 16:          return "16 años"
    if e == 17:          return "17 años"
    if 18 <= e <= 29:    return "18 a 29 años"
    if 30 <= e <= 45:    return "30 a 45 años"
    if 46 <= e <= 60:    return "46 a 60 años"
    if 61 <= e <= 75:    return "61 a 75 años"
    return "más de 75 años"  # con tilde, igual que en el Excel

# ================================
# 4) Sexo y edad en tu base
# ================================
df_hogar["sexo_std"]   = df_hogar["sexo"].apply(normalizar_sexo_valor)
df_hogar["edad_label"] = pd.to_numeric(df_hogar["años_cumplidos"], errors="coerce").apply(edad_a_tramo)

# ================================
# -------------------------
# 5) Merge exacto por etiquetas (edad_label, sexo)
# -------------------------
df_hogar = df_hogar.merge(
    tabla_long.rename(columns={"sexo_tabla":"sexo_ref"}),
    left_on=["edad_label", "sexo_std"],
    right_on=["edad_label", "sexo_ref"],
    how="left",
    suffixes=("", "_tabla")   # conserva tus columnas originales
).drop(columns=["sexo_ref"], errors="ignore")

# ================================
# 6) Suma por hogar (preserva NaN)
# ================================
df_hogar["ad_equiv_hogar3"] = (
    df_hogar.groupby(["codusu","nro_hogar"])["adulto_equiv"]
            .transform(lambda s: s.sum(min_count=1))
)
# ================================
# 7) Diagnóstico rápido
# ================================
print("Con coeficiente asignado:", df_hogar["adulto_equiv"].notna().sum(), "/", len(df_hogar))
print("Valores únicos de edad_label que no machearon (si los hubiera):")
print(df_hogar.loc[df_hogar["adulto_equiv"].isna(),"edad_label"].dropna().value_counts().head(10))
print("Valores únicos de sexo_std que no machearon (si los hubiera):")
print(df_hogar.loc[df_hogar["adulto_equiv"].isna(),"sexo_std"].dropna().value_counts().head(10))

# chequeo visual
# Vista de control con el nombre correcto
print(df_hogar[[
    "codusu","nro_hogar","sexo","sexo_std",
    "años_cumplidos","edad_label","adulto_equiv","ad_equiv_hogar3"
]].head(20))

```

```python
import numpy as np
import pandas as pd

# Mapa de AÑOS TOTALES cuando el nivel está FINALIZADO.
niveles_totales = {
    1: 0,   # Sin instrucción
    2: 6,   # Primario completo
    3: 12,  # Secundario completo (6+6)
    4: 15,  # Terciario completo (12+3)  ← ajustable
    5: 17,  # Universitario completo (12+5) ← ajustable
}

def calcular_educ(row):
    # Tomo valores y los convierto a numéricos por si vienen como string/objeto
    nivel       = pd.to_numeric(row.get("CH12"), errors="coerce")
    finalizo    = pd.to_numeric(row.get("CH13"), errors="coerce")
    ultimo_anio = pd.to_numeric(row.get("CH14"), errors="coerce")

    # Si no hay info de nivel → NaN
    if pd.isna(nivel):
        return np.nan

    nivel = int(nivel)

    # Caso sin instrucción
    if nivel == 1:
        return 0.0

    # Si finalizó el nivel → asigno años totales del nivel 
    if finalizo == 1:
        return float(niveles_totales.get(nivel, np.nan))

    # Si NO finalizó → base años previos + último año aprobado
    if finalizo == 2:
        if pd.isna(ultimo_anio):
            return np.nan
        ua = float(ultimo_anio)

        if nivel == 2:          # Primario
            return ua
        elif nivel == 3:        # Secundario
            return 6 + ua
        elif nivel in (4, 5):   # Terciario / Universitario
            return 12 + ua

    # Cualquier otro caso queda como NaN
    return np.nan


df_final["educ"] = df_final.apply(calcular_educ, axis=1).astype("float64")


resumen_educ = (
    df_final["educ"]
    .describe(percentiles=[0.5])
    .rename(index={
        "mean": "Promedio",
        "std": "Desvío estándar",
        "min": "Mínimo",
        "50%": "Mediana (p50)",
        "max": "Máximo",
    })
)
print(resumen_educ)
```

```python
#BIPLOT PCA (PC1–PC2): scores + flechas de loadings 
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#  columnas y datos 
cols = ["edad", "edad2", "educ", "ingreso_total_familiar", "miembros_hogar", "horastrab"]
X = out[cols].dropna().astype(float).copy()

#  estandarización + PCA 
Z = StandardScaler().fit_transform(X)
pca = PCA(n_components=2, random_state=0)     # solo PC1 y PC2 para el biplot
scores = pca.fit_transform(Z)                 # (n_obs x 2)
expl   = pca.explained_variance_ratio_        # varianza explicada
loadings = pd.DataFrame(pca.components_.T, index=cols, columns=["PC1","PC2"])  # (p x 2)

#  escala de flechas (que entren en la nube de puntos) 
# radio aproximado de la nube en cada eje
rx = (scores[:,0].max() - scores[:,0].min()) / 2
ry = (scores[:,1].max() - scores[:,1].min()) / 2
arrow_scale = 0.9 * min(rx, ry)               # 90% del semieje menor

#  figura 
fig, ax = plt.subplots(figsize=(7.2, 6.5))

# 1) puntos (scores)
ax.scatter(scores[:,0], scores[:,1], s=10, alpha=0.35, label="Observaciones")

# 2) flechas (loadings) escaladas
for var, (lx, ly) in loadings.iterrows():
    ax.arrow(0, 0,
             lx * arrow_scale, ly * arrow_scale,
             color="crimson", lw=1.8, alpha=0.9,
             head_width=arrow_scale*0.05, head_length=arrow_scale*0.06,
             length_includes_head=True)
    # etiqueta un poco más allá de la punta
    ax.text(lx * arrow_scale * 1.12, ly * arrow_scale * 1.12,
            var, color="navy", fontsize=10, ha="center", va="center")

# 3) ejes, títulos y guía
ax.axhline(0, color="gray", lw=1); ax.axvline(0, color="gray", lw=1)
titulo = (
    f"NOA — Biplot de PCA (PC1 vs PC2)\n"
    f"Varianza explicada: PC1 {expl[0]*100:.1f}%, PC2 {expl[1]*100:.1f}%"
)
ax.set_title(titulo)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.grid(True, ls="--", alpha=0.35)
ax.legend(loc="upper right", frameon=True)

# límites simétricos para que 0 quede centrado
m = max(abs(scores[:,0]).max(), abs(scores[:,1]).max(), arrow_scale*1.25)
ax.set_xlim(-m, m); ax.set_ylim(-m, m)
ax.set_aspect("equal", "box")

plt.tight_layout()

#  guardar 
outdir = FIGDIR if "FIGDIR" in globals() else "figs_tp2"
os.makedirs(outdir, exist_ok=True)
figpath = os.path.join(outdir, "03_pca_biplot_scores_loadings.png")
plt.savefig(figpath, dpi=300, bbox_inches="tight", facecolor="white")
plt.show()

print("Figura guardada en:", figpath)

#  tabla de loadings
display(loadings.sort_values("PC1", key=lambda s: s.abs(), ascending=False).round(2))
display(loadings.sort_values("PC2", key=lambda s: s.abs(), ascending=False).round(2))
```

```python

# B. Clúster k-medias (edad + ITF)
# Base: out (NOA, jefes). df_final tiene 'pobre'

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#  carpeta de salida 
OUTDIR = FIGDIR if 'FIGDIR' in globals() else 'figs_tp2'
os.makedirs(OUTDIR, exist_ok=True)

#  1) base con 2 variables 
itf_col = "ingreso_total_familiar" if "ingreso_total_familiar" in out.columns else "ingreso_total_familiar_2025"
base = out[["edad", itf_col]].dropna().astype(float).copy()
base.columns = ["edad", "itf"]

# traigo 'pobre' (0/1) alineado al índice de 'base'
pobre_col = next(c for c in ["pobre", "POBRE", "Pobre"] if c in df_final.columns)
y_true = pd.to_numeric(df_final.loc[base.index, pobre_col], errors="coerce").astype("Int64")

#  2) estandarizo (z-score) 
scaler = StandardScaler()
Z = scaler.fit_transform(base)  # [edad_z, itf_z]

# ejes desde zero (buena visualzicion de datos)
def ejes_desde_cero(ax):
    ax.set_xlim(left=0)   # <-- arranca X en 0
    ax.set_ylim(bottom=0) # <-- arranca Y en 0

#  3) función para correr y graficar cada k 
def corre_y_grafica_k(k: int):
    km = KMeans(n_clusters=k, n_init=20, random_state=0)
    labels = km.fit_predict(Z)
    cents = scaler.inverse_transform(km.cluster_centers_)  # centroides en escala original

    fig, ax = plt.subplots(figsize=(6.8, 5.4))
    ax.scatter(base["edad"], base["itf"], c=labels, s=12, alpha=0.55)
    ax.scatter(cents[:, 0], cents[:, 1], marker="X", s=160, edgecolor="k", label="Centroides")
    ax.set_title(f"NOA — k-medias con k={k} (Edad vs ITF)")
    ax.set_xlabel("Edad")
    ax.set_ylabel("Ingreso total familiar (ITF)")
    ax.legend(loc="best", frameon=True)

    ejes_desde_cero(ax)   # <-- forzar origen en 0,0

    plt.tight_layout()
    fig.savefig(os.path.join(OUTDIR, f"05_kmedias_k{k}_edad_itf_noa.png"),
                dpi=300, bbox_inches="tight", facecolor="white")
    plt.show()

    sil = silhouette_score(Z, labels)
    print(f"Silhouette (k={k}): {sil:.3f}")

    if k == 2 and y_true.notna().any():
        mapa = {}
        for c in np.unique(labels):
            y_c = y_true[labels == c].dropna()
            mapa[c] = int((y_c.mean() >= 0.5)) if len(y_c) else 0
        y_pred = np.array([mapa[c] for c in labels], dtype=int)
        acc = (y_pred == y_true.to_numpy(dtype=int)).mean()
        print(f"Alineación k=2 vs condición de pobreza (mayoría por clúster): exactitud = {acc:.2%}")

        cm = pd.crosstab(pd.Series(y_true, name="Pobre (real)"),
                         pd.Series(y_pred, name="Clúster → Pobre (pred.)"))
        print(cm)

#  4) correr k = 2, 4, 10 
for k in (2, 4, 10):
    corre_y_grafica_k(k)

#  5) comparación visual específica para k=2 vs pobreza 
from matplotlib.gridspec import GridSpec

km2 = KMeans(n_clusters=2, n_init=20, random_state=0).fit(Z)
lab2 = km2.labels_

fig = plt.figure(figsize=(11, 4.6))
gs = GridSpec(1, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax1.scatter(base["edad"], base["itf"], c=lab2, s=12, alpha=0.55)
ax1.set_title("k=2 — Clústeres (Edad vs ITF)")
ax1.set_xlabel("Edad")
ax1.set_ylabel("Ingreso total familiar (ITF)")
ejes_desde_cero(ax1)  # <-- origen 0,0

ax2 = fig.add_subplot(gs[0, 1])
ax2.scatter(base["edad"], base["itf"], c=y_true.to_numpy(dtype=float), s=12, alpha=0.55)
ax2.set_title("Color por condición de pobreza (0 = no, 1 = sí)")
ax2.set_xlabel("Edad")
ax2.set_ylabel("Ingreso total familiar (ITF)")
ejes_desde_cero(ax2)  # <-- origen 0,0

plt.tight_layout()
fig.savefig(os.path.join(OUTDIR, "05b_kmedias_k2_clusters_vs_pobreza_noa.png"),
            dpi=300, bbox_inches="tight", facecolor="white")
plt.show()
```

```python

# Elbow (disimilitud) k = 1..10
# Variables: edad + ITF (escala z)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# 1) base 2D
itf_col = "ingreso_total_familiar" if "ingreso_total_familiar" in out.columns else "ingreso_total_familiar_2025"
base = out[["edad", itf_col]].dropna().astype(float).copy()
base.columns = ["edad", "itf"]

# 2) estandarizar
Z = StandardScaler().fit_transform(base)

# 3) SSE (inertia) y Silhouette (opcional)
ks = range(1, 11)
sse = []
sil = []
for k in ks:
    km = KMeans(n_clusters=k, n_init=20, random_state=0)
    labs = km.fit_predict(Z)
    sse.append(km.inertia_)                         # disimilitud (SSE)
    if k >= 2:
        sil.append(silhouette_score(Z, labs))       # compacidad/separación

# 4) Gráfico del codo (Elbow)
plt.figure(figsize=(6.4, 4))
plt.plot(ks, sse, marker="o")
plt.xticks(ks)
plt.xlim(left=0)     # <-- eje X arranca en 0
plt.ylim(bottom=0)   # <-- eje Y arranca en 0
plt.xlabel("Número de clústeres (k)")
plt.ylabel("Inercia intra-clúster (SSE)")
plt.title("NOA — Método del codo (Edad + ITF, k-medias)")
plt.tight_layout()
figsave("06_elbow_noa.png")
plt.show()

# 5) Gráfico del índice de silueta para k ≥ 2 — estético y con ejes desde 0
xs = list(range(2, 11))
best_idx = int(np.argmax(sil))
best_k   = xs[best_idx]
best_val = float(sil[best_idx])

fig, ax = plt.subplots(figsize=(7.2, 4.4))

# curva principal
ax.plot(xs, sil, marker="o", linewidth=2.2, markersize=6)

# relleno suave
ax.fill_between(xs, sil, 0, alpha=0.08)

# punto y línea vertical del mejor k
ax.scatter([best_k], [best_val], s=60, zorder=3)
ax.axvline(best_k, linestyle="--", linewidth=1, alpha=0.6)

# ejes DESDE 0
ax.set_xlim(0, 10.5)
ax.set_ylim(0, max(0.01, max(sil)*1.12))

# ticks y grilla
ax.set_xticks(range(0, 11, 1))
ax.grid(True, linestyle="--", alpha=0.35)
ax.set_axisbelow(True)

# etiquetas y título
ax.set_xlabel("Número de clústeres (k)")
ax.set_ylabel("Índice de silueta")
ax.set_title(f"NOA — Índice de silueta (Edad + ITF, k-medias)\n"
             f"Mejor k ≈ {best_k}  (s = {best_val:.3f})")

plt.tight_layout()

# guardado 
try:
    figsave("07_silhouette_noa.png")
except NameError:
    import os
    outdir = FIGDIR if 'FIGDIR' in globals() else 'figs_tp2'
    os.makedirs(outdir, exist_ok=True)
    fig.savefig(os.path.join(outdir, "07_silhouette_noa.png"),
                dpi=300, bbox_inches="tight", facecolor="white")

plt.show()
```

```python
# ===== Clustering jerárquico (Ward) 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# columnas 
itf_col = "ingreso_total_familiar" if "ingreso_total_familiar" in out.columns else "ingreso_total_familiar_2025"
cols = ["edad", "edad2", "educ", itf_col, "miembros_hogar", "horastrab"]

# 1) datos numéricos sin NAs + z-score
X = out[cols].dropna().astype(float).copy()
Z = StandardScaler().fit_transform(X)

# 2) jerárquico (Ward usa distancia euclídea y minimiza varianza intra)
Z_link = linkage(Z, method="ward")

# 3A) Dendrograma (completo)  
plt.figure(figsize=(10, 4.6))
dendrogram(Z_link, no_labels=True, color_threshold=None)
plt.title("NOA — Dendrograma (Ward, z-score)")
plt.ylabel("Distancia (inercia agregada)")
plt.tight_layout()
figsave("08_silhouette_noa.png")
plt.show()

# 3B) Dendrograma truncado : últimas 30 fusiones
plt.figure(figsize=(10, 4.6))
dendrogram(
    Z_link, truncate_mode="lastp", p=30, show_contracted=True,
    leaf_rotation=45, leaf_font_size=9, color_threshold=None
)
plt.title("NOA — Dendrograma truncado (últimas 30 fusiones)")
plt.ylabel("Distancia")
plt.tight_layout()
figsave("09_silhouette_noa.png")
plt.show()

# 4) elegimos k
k = 4  
labels_k = fcluster(Z_link, k, criterion="maxclust")
print(pd.Series(labels_k).value_counts().sort_index())

# 5) Edad vs ITF coloreado por clúster jerárquico
fig, ax = plt.subplots(figsize=(6.8, 5.2))
ax.scatter(X["edad"], X[itf_col], c=labels_k, s=12, alpha=0.55)

ax.set_xlabel("Edad")
ax.set_ylabel("Ingreso total familiar (ITF)")
ax.set_title(f"NOA — Clustering jerárquico (Ward), k = {k}")

# que empiece en 0
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)

plt.tight_layout()
fig.savefig(os.path.join(OUTDIR, "10_hclust_edad_itf_noa.png"),
            dpi=300, bbox_inches="tight", facecolor="white")
plt.show()
```
